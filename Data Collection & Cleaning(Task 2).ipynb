```python
#Drops a column
df=df.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
```

Remove Duplicates

```python
#Remove duplicates
df =df.drop_duplicates()
```

Handle Missing Values

```python
df.isnull().sum()
```

Parse and Validate Timestamps

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

```python
df.head(2)
```

Clean Text Data (Text Column)

```python
df['Text']=df['Text'].str.strip()
```

```python
df['Text']=df['Text'].str.replace(r'\s+', ' ',regex=True)
```

Clean Categorical Columns (Strip Whitespaces)

```python
cols_to_strip = ['Sentiment','User','Platform','Hashtags','Country']
df[cols_to_strip] = df[cols_to_strip].apply(lambda x: x.str.strip())
```

Fix Data Types for Numerical Columns

```python
df['Retweets'] = df['Retweets'].astype(int)
df['Likes'] = df['Likes'].astype(int)
```

Detect and Handle Outliers

```python
num_cols = ['Retweets', 'Likes', 'Year','Month','Day','Hour']

for col in num_cols:
    Q1 =df[col].quantile(0.25)
    Q3 =df[col].quantile(0.75)
    IQR= Q3-Q1
    lower =Q1 - 1.5 * IQR
    upper=Q3 + 1.5 * IQR

    df[f'{col}_Outlier'] = ((df[col] < lower) | (df[col] > upper)).astype(int)

df[[col for col in df.columns if 'Outlier' in col]].sum()
```

Prepare Numerical Data for Scaling

```python
X_num = df[num_cols].copy()
print(X_num.describe())
```

Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler_z = StandardScaler()
X_z = scaler_z.fit_transform(X_num)
df_z = pd.DataFrame(X_z, columns=[f'{col}_z' for col in num_cols])

# Stats for Retweets
retweets_mean = X_num['Retweets'].mean()
retweets_std = X_num['Retweets'].std()
retweets_z_mean = df_z['Retweets_z'].mean()
retweets_z_std = df_z['Retweets_z'].std()

# Example: First value (15)
first_retweet = X_num['Retweets'].iloc[0]
first_z = (first_retweet - retweets_mean) / retweets_std

print(f"Retweets Z-score stats: Mean={retweets_z_mean:.3f}, Std={retweets_z_std:.3f}")
print(f"First Z-score: {first_z:.3f}")
```

Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler_mm = MinMaxScaler()
X_mm = scaler_mm.fit_transform(X_num)
df_mm = pd.DataFrame(X_mm, columns=[f'{col}_mm' for col in num_cols])

# Stats for Retweets
retweets_min = X_num['Retweets'].min()
retweets_max = X_num['Retweets'].max()
retweets_mm_min = df_mm['Retweets_mm'].min()
retweets_mm_max = df_mm['Retweets_mm'].max()

# Example: First value
first_mm = (first_retweet - retweets_min) / (retweets_max - retweets_min)

print(f"Retweets Min-Max stats: Min={retweets_mm_min:.3f}, Max={retweets_mm_max:.3f}")
print(f"First Min-Max: {first_mm:.3f}")
```
